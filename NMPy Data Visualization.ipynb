```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x = np.linspace(-5,5,1000)
x.shape
mod_x = abs(x)
plt.plot(x,mod_x,"r--")
plt.show()
x = np.linspace(-5,5)
x_sq = x**2
x_cube = x**3
x_sqrt = np.sqrt(x)
x_exp  = np.exp(x)
plt.plot(x,x,'r:',label='Line',lw=3)
#plt.plot(x,x_sqrt,'g--',label="Sqrt",lw=3)
plt.plot(x,x_sq,'g--',label="Squre",lw=3)
plt.plot(x,x_cube,'b-.',label="Cube",lw=3)
plt.plot(x,x_exp,'m-',label="Exp",lw=2)
plt.legend()
plt.show()



```

```python
x = np.arange(1,100,13)
plt.plot(x,x+10,'o--',label="legend")
plt.title("Hello World to Matplot",fontsize=10)
plt.xlabel("hello i am x-axis")
plt.ylabel("hello i am y-axis")
plt.xticks(x)
plt.yticks(x+10)
plt.grid()
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
angle = np.arange(0,361,15)
rad = np.deg2rad(angle)
sin = np.sin(rad)
cos = np.cos(rad)
tan=np.tan(rad)
plt.plot(angle,sin,label='sin')

plt.plot(angle,cos,label='cos')
plt.plot(angle,tan,label='tan')

plt.xticks(range(-30,391,15))
plt.yticks([-2,-1,-0.5,0,0.5,1,2])
plt.xlabel("Angles",fontsize=20)
plt.ylabel("Values",fontsize=20)
plt.title("Sin Vs. Cos Vs Tan",fontsize=30)
plt.legend()
plt.grid()
plt.savefig("sin_cos.png")
plt.show()
```
