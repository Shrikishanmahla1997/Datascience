```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x=np.linspace(-1,1,1000)
x_mod=abs(x)
x=x
x_exp=np.exp(x)
x_sq=x**2
x_cube=x**3
plt.plot(x,x_mod,"b--",label="Mod",lw=3)
plt.plot(x,x,"y--",label="line",lw=3)
plt.plot(x,x_sq,"r--",label="Sqrt",lw=3)
plt.plot(x,x_cube,"g--",label="cube",lw=5)
plt.plot(x,x_exp,"m--",label="Expo",lw=3)
plt.legend()
plt.show()
```
